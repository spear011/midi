```python
def open_midi(midi_path):
    """
    Open the MIDI file and return the score.

    Returns:
        Score object.
    """
    score = converter.parse(midi_path)

    # Remove percussion staff.
    for part in score.parts:
        if 'Percussion' in part.partName or 'Drum' in part.partName:
            score.remove(part)
    return score
```
